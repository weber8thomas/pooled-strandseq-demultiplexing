```python
# list the content of this: https://ftp.1000genomes.ebi.ac.uk/vol1/ftp/data_collections/HGSVC2/working/20200422_Strandseq 
from ftplib import FTP

def list_ftp_directory(url, path):
    # Parse the URL to extract the FTP server domain
    ftp_server = url.split('/')[2]

    # Connect to the FTP server
    with FTP(ftp_server) as ftp:
        # Login as anonymous
        ftp.login()

        # Change to the specified directory
        ftp.cwd(path)

        # List contents of the directory and print them
        contents = ftp.nlst()  # Alternatively, use ftp.dir() to get more detailed listings
        return contents




# URL and Path for the FTP directory
url = 'https://ftp.1000genomes.ebi.ac.uk/vol1/ftp/data_collections/HGSVC2/working/20200422_Strandseq'
path = '/vol1/ftp/data_collections/HGSVC2/working/20200422_Strandseq'

# Call the function to list directory contents
contents = list_ftp_directory(url, path)
print(contents)
contents_dict = {}
for item in contents:
    # if item is folder (no extension)
    if '.' not in item:
        # print(item)
        # Pick the first file in the item folder
        item_path = path + '/' + item
        item_contents = list_ftp_directory(url, item_path)
        # print(item_contents)
        item_prefixes = set([x.split("PE20")[0] for x in item_contents])
        if len(item_prefixes) == 1:
            # print(item_prefixes)
            # Get shortest substring before PE20 from all the files in the folder
            
            first_file = item_contents[0].split('/')[0].split("PE20")[0]
    #         # print(first_file)
        else:
            print("Multiple prefixes found")
            first_file = None
        contents_dict[item[2:]] = first_file
contents_dict
        
```
